### Blackjack Game

Creating a simple text-based BlackJack game.

- The game needs to have one player versus an automated dealer.
- The player can stand or hit.
- The player must be able to pick their betting amount.
- Keep track of the player's total money.
- Give alert to the player of wins, losses, or busts, etc...
- Using OOP and classes in our game.

```python
# import statements

from random import shuffle
from random import choice


global playing
playing=True

# Class definitions:

# Class card with attributes number and sign, and method to return value of card according to the blackjack rules

class card:
    
    def __init__(self,number,sign):
        
        self.number=number
        self.sign=sign
        
    def value(self):
        
        if self.number in ['queen','jack','king']:
            value=10
        elif self.number =='ace':
            value=11
        else:
            value=self.number
        return value
    
# The class table is a child of the card class and contains methods: division,hit and shuffle.

class desk(card):
    
    def __init__(self,card):
        self.card=card
        
    def division(self):
        div=self.card[:2]
        for card in div:
            self.card.remove(card)   
        return div
    
    def hit(self):
        div=self.card[:1]
        for card in div:
            self.card.remove(card)   
        return div
    
    def shuffle(self):
        
        shuffle(self.card)
        

# Class diller is child class of desk class, for its initialization it is necessary to give a list of cards

class diller(desk):
    
    def __init__(self,name,card_list=None):
        
        if card_list is None:
            card_list = [] # create a new list
        self.card_list= card_list
        
        self.name=name
        
        self.card_list=card_list
        
    def add_division(self,division_list):
        for card in division_list:
            self.card_list.append(card)
            
    def result(self):
        numbers=[card.value() for card in self.card_list]
        
        while numbers.count(11)>=2:
            numbers.remove(11)
            numbers.append(1)   
        suma=0
        for n in numbers:
            suma=suma+n
        
        return suma

# Class player has two parent classes diller and desk, and can use all their attributes and methods

class player(diller,desk):
    
    def __init__(self,name,card_list=None,balance=100):
        
        self.balance=balance
        self.name=name
        if card_list is None:
            card_list = [] # create a new list
        self.card_list= card_list
       
        
    def change_balnce(self,value):
        
        if value>0:
            self.balance+=2*value
        else:
            self.balance+=value
            
# Once we make a classes let's define a function that is needed for playing the game.

def deck_maker():
    
    ''' 
    A function generate deck of 52 cards,every card is instance of card class card has two attributes,
    value and sign one method that gives back value of card for calculation.
    '''
    deck=[]
    j=0
    for s in ['heart','clube','diamond','spade']:
        for i in [2,3,4,5,6,7,8,9,10,'ace','queen','jack','king']:
            j+=1
            k="k"+str(j)
            k=card(i,s)
            deck.append(k)
    return deck

def player_bet(player):
    
    '''
    Player inputs bet, if  his current balance is less than inputted, 
    the function do not accept players bet, but asks  to input  a new bet
    
    '''
    
    bet=int(input("{}, please insert Your bet=".format(player.name)))
    while player.balance<bet:
        print('Your curren balanc is {})'.format(player.balance))
        bet=int(input("{}, please insert Your bet=".format(player.name)))
        
    return bet


def display_all(player):
    
    '''
    A function that displays all cards in card list of player
    '''
    
    print()
    print("{} cards are:".format(player.name),end=" ")
    for card in player.card_list:
        print("{}:{}".format(card.sign,card.number),end=", ")
    print("\nScore is {}".format(player.result()))
    print()

def display_diller(diller):
    
    '''
    A funcion that displays only one card of diller according to the rule
    
    '''
    
    card=diller.card_list[1]
    print("\n{} cards are:{}:{}".format(diller.name,card.sign,card.number))
    card=diller.card_list[1]
    print()
    
    
def hit_player(player,desk):
    
    '''
    A function that asks player to hit or stop, and if player hits 
    call method of player to add card in player list
    '''
    hint=input("Please, insert h for hit or s for stand = ")
    if hint.lower()=='h':
        player.add_division(desk.hit())
    else:
        print("Score is {}".format(player.result()))
    return hint.lower()



def hit_diller(diller,desk):
    
    '''
    A function that, according to the rules of the game, hits the dealer as long as the score is below 17,
    with method add_division 
    '''
    
    while diller.result()<17:
        diller.add_division(desk.hit())
        

def player_bust(player):
    
    '''
    A function that checks if the sum of the player's cards is greater than 21, which means that he has lost
    '''
    flag=False
    if player.result()>21:
        print("Sorry {}, You busts".format(player.name))
        flag=True
    return flag
    

def player_bust_check(player):
    
    '''
    A function that checks if the sum of the diller's cards is greater than 21, which means that player has won
    '''
    flag=False
    if player.result()>21:
        flag=True
    return flag



def black_jack(player,diller):
    '''
    A function that checks if player or diller got black_jack
    '''
    
    flag=False
    
    if player.result()==21:
        print("Black Jack,{} has won ".format(player.name))
        flag=True
    elif diller.result()==21:
        print("Black Jack,{} has won ".format(diller.name))
        flag=True
    return flag
    
        
          
def winner(player,diller,bet):
    '''
    A function that checks who is winner
    '''
    
    if player.result() > diller.result():
        player.change_balnce(bet)
        print('{} has won !'.format(player.name))
    elif player.result()< diller.result():
        player.change_balnce(-bet)
        print('{} has won !'.format(diller.name))
    else:
        print('Nobody won!!!')


        
def start_new_game(player,diller,bet):
    
    '''
    A function for starting a  new game, if it is possible, depending on available funds
    '''
    if player.balance>2:
        start=input("If You wont to start new game, insert Y/N=")
        if start.lower()=='y':
            player.card_list=[]
            diller.card_list=[]
            start_game(player)
        else:
            print("Thank You, see You soon!!!")
        
    else:
        print('Dear player, you do not have any available funds on your account,game is over!!!\n If you wish to play again, please deposit some funds on your account.')
    
    
```

```python
def start_game(player):
    
    '''
    1. Game starts with argument (player instance of class player)
    2. By default player has balance $100
    3. Initialization  diller as inctance of  diller class
    4. Initialization  desk as inctance of  desk class
       (for inizialization desk we need list of 52 card from card class)
    5. Calling method add_division that put two cards in card_list of player and diller
    6. Check if black_jack has occurred
         if that happens, the game is over
         if not, then the player decides whether he wants to be hit or not
         if a player selects a hit, the add_card method from the player class is called and the result is checked
         if result is over 21 playing become False and game is over
         if result of player is below 21, function hit diller is called,  if the result of diller's cards is over 17 or over 21 playing is False 
         and  we calling function hit player and if answer is y add new card in card_list of player and check if player is bust.
         if player choose n or player bust playing =False
         after that we just check who has won
         if player won we add bet amount to his balance, if he lose we take bet mamount from his balance
    7. Player can chose if he wants to play again, conditions are: 
       if he has balance less the $2 game is over
       if he put bet greater than his balance, he wouldn't be able to play, he wolud be asked to enter a new correct value, or to deposit samo funds inti his account 
       and the new game can start.
    '''
    player1=player
    playing=True
    #formating desk and diller
    bet=0
    diller1=diller("Diller")
    desk1=desk(deck_maker())
    desk1.shuffle()
    
    #player inserts bet
    while not( 2 < bet < 500 ):
        bet=player_bet(player1)
    
    #game starts with division of two cards
    player1.add_division(desk1.division())
    diller1.add_division(desk1.division())
    display_all(player1)
    display_diller(diller1)
    
    #checking blackjack
    
    if black_jack(player1,diller1)==False:
        if hit_player(player1,desk1)=="h":
            if player_bust(player1)==False:
                hit_diller(diller1,desk1)
                if player_bust(diller1):
                    playing=False     
            else:
                playing=False
        else:
            hit_diller(diller1,desk=desk1)
            player_bust(diller1)
            playing=False
    else:
        playing=False
        
            
    while playing:
    
        display_all(player1)
        display_all(diller1)
    
        if (hit_player(player1,desk1)).lower()=="s" or player_bust(player1) or player_bust(diller1):
            playing=False
    
    win=""
    
    if player_bust_check(player1):
        win="diller"
        print()
        print("{} has won!".format(diller1.name))
        display_all(diller1)
        player1.change_balnce(-bet)
        
    elif player_bust_check(diller1):
        win="player"
        print()
        print("{} has won!".format(player1.name))
        display_all(diller1)
        player1.change_balnce(bet)
        
    if win not in ["player","diller"]:
        winner(player1,diller1,bet=bet)
        
    start_new_game(player1,diller1,bet=bet)
    playing=False
        
```

```python
def main():
    
    '''
    Start game
    '''
    print('Welcome to BlackJack! Get as close to 21 as you can without going over!\nAces count as 1 or 11.')
    
    
    #player inserts name
    
    name=""
    while len(name)<2:
        name=input("Please, enter your name: ")
    
    player1=player(name)
    
    start_game(player1)
    
if __name__ == '__main__':
    main()
```
